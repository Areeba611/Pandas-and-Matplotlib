# Section 5: Sorting and Ranking
1. Sort values by one or more columns
2. Rank entries within groups


```python
df.sort_values('cases', ascending=False)
```

```python
df.sort_values(by=['States', 'cases'], ascending=[True, False])
```

```python
df['case_rank'] = df.groupby('States')['cases'].rank(method='dense', ascending=False)
```

```python
df_ranked = df[['States', 'Country', 'cases', 'case_rank']]
print(df_ranked.sort_values(by=['States', 'case_rank']).head(10))
```

# Section 6: Creating and Modifying Columns
1. Create new column using existing ones
2. Use apply() for custom functions


```python
df['death_rate'] = (df['fips'] + df['cases']) 
```

```python
df
```

```python
def case_severity(row):                                                # Custom Function
    if row['cases'] > 100000:
        return 'High'
    elif row['cases'] > 50000:
        return 'Medium'
    else:
        return 'Low'
```

```python
df['severity_level'] = df.apply(case_severity, axis=1)
```

```python
df[['States', 'Country', 'cases', 'severity_level']]
```

# Section 7: Visualization (using matplotlib)
1. Plot line graph.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Covid-19_us-counties.csv")
df['date'] = pd.to_datetime(df['date'])                     # Important because the default data type of date is object(string) 
country_data = df[df['county'] == 'Maricopa']
plt.plot(country_data['date'], country_data['cases'], label='Cases', color='green')
plt.title('COVID Cases in Maricopa')
plt.xlabel('Dates')
plt.ylabel('Number of Cases')
plt.grid(True, color='r')
plt.show()
```
